```python
driver = webdriver.Chrome()
url = "https://www.jobkorea.co.kr/Recruit/GI_Read/46099393?rPageCode=SL&logpath=21&sn=6"
driver.get(url)

try:
    # iframe 전환(전환하지 않으면 접근 불가함)
    WebDriverWait(driver, 10).until(
        EC.frame_to_be_available_and_switch_to_it((By.ID, "gib_frame"))
    )
    print("iframe 전환 완료")

    # iframe 안의 내용을 가져오기
    content_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "content_sec"))
    )
    
    # 텍스트 가져오기
    content_text = content_element.text
    print("iframe 내 텍스트 내용:")
    print(content_text)

    # HTML 전체를 가져오고 싶다면
    content_html = content_element.get_attribute("outerHTML")
    print("iframe 내 HTML 내용:")
    print(content_html)

except Exception as e:
    print(f"오류 발생: {e}")


```
